### 문제 출처 - 리트 코드 [125번 문제 Valid Palindrome](https://leetcode.com/problems/valid-palindrome)

### 내 풀이

```python
from typing import *
class Solution:
    def isPalindrome(self, s: str) -> bool:
        if len(s) <= 1:
            return False
        s = s.lower()
        strs = []
        for char in s:
            if char.isalnum():
                strs.append(char)
        for i in range(len(strs)//2):
            if strs[i] != strs[-i-1]:
                return False
        return True

sol = Solution()
s = 'A man, a plan, a canal: Panama'
s1 = 'B'
print(f'{s} is palindrome?  {sol.isPalindrome(s)}')
print(f'{s1} is palindrome?  {sol.isPalindrome(s1)}')
```

```python
from typing import *
class Solution_1:
    def isPalindrome(self, s: str) -> bool:
        strs = []
        for char in s:
            if char.isalnum():
                strs.append(char.lower())

        # 팰린드롬 여부 판별
        while len(strs) > 1:
            if strs.pop(0) != strs.pop():
                return False

        return True
```

### 풀이 #2

```python

import collections
from typing import Deque


class Solution:
    def isPalindrome(self, s: str) -> bool:
        # 자료형 데크로 선언
        strs: Deque = collections.deque()

        for char in s:
            if char.isalnum():
                strs.append(char.lower())

        while len(strs) > 1:
            if strs.popleft() != strs.pop():
                return False

        return True
```

### 풀이 #3

```python

import re


class Solution:
    def isPalindrome(self, s: str) -> bool:
        s = s.lower()
        # 정규식으로 불필요한 문자 필터링
        s = re.sub('[^a-z0-9]', '', s)

        return s == s[::-1]  # 슬라이싱
```
