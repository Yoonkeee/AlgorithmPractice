```python
from typing import *
import collections
class Solution():
    def func(self, strs):
        sorted_dict:dict = collections.defaultdict(list)
        for word in strs:
            sorted_word = ''.join(sorted(word))
            sorted_dict[sorted_word].append(word)  # ì—†ëŠ” key ì°¸ì¡°ì‹œ defaultê°€ listì´ë¯€ë¡œ append ê°€ëŠ¥
        return list(sorted_dict.values())
    
strs = ['eat', 'tea', 'tan', 'ate', 'nat', 'bat']
group_anagrams = Solution()
group = group_anagrams.func(strs)
print(group)
```

```python
from typing import *
import collections
class Solution():
    def func(self, strs):
        anagrams = collections.defaultdict(list)
        
        for word in strs:
            anagrams[''.join(sorted(word))].append(word)
        return anagrams.values()
    
strs = ['eat', 'tea', 'tan', 'ate', 'nat', 'bat']
group_anagrams = Solution()
group = group_anagrams.func(strs)
print(group)
```
