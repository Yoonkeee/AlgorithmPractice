```python
from typing import *
import collections
class Solution():
    def func(self, input_data):
        somevar = input_data

    def __init__(self):
        self.q = collections.deque()
        
    def push(self, x: int) -> None:
        self.q.append(x)
        for _ in range(len(self.q)-1):
            self.q.append(self.q.popleft())
            
    def pop(self) -> int:
        return self.q.popleft()
        
    def top(self) -> int:
        if len(self.q) > 0:
            return self.q[0]
        

    def empty(self) -> bool:
        return len(self.q) == 0
```
