### 문제 출처 - 리트 코드 [344번 문제 Reverse String](https://leetcode.com/problems/reverse-string)

```python
from typing import *
class Solution():
    def reverse_string(self, s: str) -> None:
        temp = []
        for i in range(len(s)-1, -1, -1):
            temp.append(s[i])
        for i in range(len(s)):
            s[i] = temp[i]

reverse = Solution()
s = ['h', 'e', 'l', 'l', 'o']
reverse.reverse_string(s)
print(s)
```

# 풀이#1 - 투 포인터를 이용한 스왑

```python
from typing import List


class Solution:
    def reverseString(self, s: List[str]) -> None:
        left, right = 0, len(s) - 1
        while left < right:
            s[left], s[right] = s[right], s[left]
            left += 1
            right -= 1
```

# 풀이#2 - 파이썬다운 방식

```python
from typing import List


class Solution:
    def reverseString(self, s: List[str]) -> None:
        s.reverse()
```
