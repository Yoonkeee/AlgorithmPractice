# 1. íˆ¬ í¬ì¸í„°ë¥¼ ìµœëŒ€ë¡œ ì´ë™

```python
from typing import *
class Solution():
    def func(self, input_data):
        height = input_data
        PRINT_LOG = True
        if not height:
            return 0
        
        volume = 0
        left, right = 0, len(height) - 1
        left_max, right_max = height[left], height[right]
        
        if PRINT_LOG:
            while left < right:
                print(f'\nleft : {left}, left_max : {left_max}, right : {right}, right_max : {right_max}')
                left_max, right_max = max(height[left], left_max), max(height[right], right_max)
                print(f'left_max=max({height[left]}, {left_max}) = {max(height[left], left_max)}', 
                      f', right_max=max({height[right]}, {right_max}) = {max(height[right], right_max)}')
                if left_max <= right_max:
                    print(f'volume {volume} += left_max {left_max} - height[left] {height[left]}')
                    volume += left_max - height[left]
                    print(f'left : {left} += 1')
                    left += 1
                else:
                    print(f'volume {volume} += right_max {right_max} - height[right] {height[right]}')
                    volume += right_max - height[right]
                    print(f'right : {right} -= 1')
                    right -= 1
        else:
            while left < right:
                left_max, right_max = max(height[left], left_max), max(height[right], right_max)
                if left_max <= right_max:
                    volume += left_max - height[left]
                    left += 1
                else:
                    volume += right_max - height[right]
                    right -= 1
        
        return volume
    

input_data = [[0,1,0,2,1,0,1,3,2,1,2,1] , [4,2,0,3,2,5]]
output_data = [6, 9]
sol = Solution()
for i in range(len(input_data)):
    print('input  :', input_data[i])
    print('result :', sol.func(input_data[i]))
    print('output :', output_data[i])
    print('')
```

```python
from typing import *
class Solution():
    def func(self, input_data):
        height:List[int] = input_data
        stack = []
        volume = 0
        PRINT_LOG = True
        
        if PRINT_LOG:
            for i in range(len(height)):
                print(f'i:{i}, stack:{stack}')
                # êº¾ì´ëŠ” ë¶€ë¶„ ë³€ê³¡ì ì„ ë§Œë‚˜ëŠ” ê²½ìš° ê²©ì°¨ë§Œí¼ ë¬¼ ë†’ì´ë¥¼ ì±„ìš´ë‹¤
                while stack and height[i] > height[stack[-1]]:
                    # ìŠ¤íƒì—ì„œ êº¼ë‚¸ë‹¤
                    print(f'In While top = {stack}.pop()={stack[-1]}')
                    top = stack.pop()
                    if not len(stack):
                        break
                    # ì´ì „ê³¼ì˜ ì°¨ì´ë§Œí¼ ë¬¼ ë†’ì´ ì²˜ë¦¬
                    print(f'distance = i:{i} - stack[-1]:{stack[-1]} - 1 = {i - stack[-1] - 1}')
                    distance = i - stack[-1] - 1
                    waters = min(height[i], height[stack[-1]]) - height[top]
                    print(f'waters = min(height[i], height[stack[-1]]) - height[top] = ',
                          f'min({height[i]}, {height[stack[-1]]}) - {height[top]} = {waters}')

                    volume += distance * waters
                print(f'\n{stack}.append({i})')
                stack.append(i)
        else:
            for i in range(len(height)):
                # êº¾ì´ëŠ” ë¶€ë¶„ ë³€ê³¡ì ì„ ë§Œë‚˜ëŠ” ê²½ìš° ê²©ì°¨ë§Œí¼ ë¬¼ ë†’ì´ë¥¼ ì±„ìš´ë‹¤
                while stack and height[i] > height[stack[-1]]:
                    # ìŠ¤íƒì—ì„œ êº¼ë‚¸ë‹¤
                    top = stack.pop()
                    if not len(stack):
                        break
                    # ì´ì „ê³¼ì˜ ì°¨ì´ë§Œí¼ ë¬¼ ë†’ì´ ì²˜ë¦¬
                    distance = i - stack[-1] - 1
                    waters = min(height[i], height[stack[-1]]) - height[top]

                    volume += distance * waters
                stack.append(i)
        return volume
    

input_data = [[0,1,0,2,1,0,1,3,2,1,2,1] , [4,2,0,3,2,5]]
output_data = [6, 9]
sol = Solution()
for i in range(len(input_data)):
    print('input  :', input_data[i])
    print('result :', sol.func(input_data[i]))
    print('output :', output_data[i])
    print('')
```
