```python
from typing import *
class WRONG_Solution():
    def func(self, data:List[int]) -> List[int]:
        nums, target = data[0], data[1]
        left, right = 0, len(nums)-1
        while nums[left] + nums[right] != target and left < right:
            if nums[left] + nums[right] < target:
                left += 1
            elif nums[left] + nums[right] > target:
                right -= 1
        return left, right
        # nums_map = {key=nums : value=index}
        # nums_map[target - num] = index
        # n in dict = key ì¡°íšŒ
        

input_data = [[[2,7,11,15], 9] , [[3,2,4], 6] , [[3,4,3] , 6]]
sol = WRONG_Solution()
for data in input_data:
    print(sol.func(data))
```

# ì¡°íšŒ êµ¬ì¡° ê°œì„ 

```python
from typing import *
class Solution():
    def func(self, data:List[int]) -> List[int]:
        nums, target = data[0], data[1]
        nums_map = {}
        for i, num in enumerate(nums):
            if target - num in nums_map:
                return [nums_map[target - num], i]
            nums_map[num] = i
        # nums_map = {key=nums : value=index}
        # nums_map[target - num] = index
        # n in dict = key ì¡°íšŒ
        

input_data = [[[2,7,11,15], 9] , [[3,2,4], 6] , [[3,4,3] , 6]]
sol = Solution()
for data in input_data:
    print(sol.func(data))
```

```python
from typing import *
class Solution():
    def func(self, data:List[int]) -> List[int]:
        nums, target = data[0], data[1]
        nums_map = {}
        for i, n in enumerate(nums):
            nums_map[n] = i
        # nums_map = {key=nums : value=index}
        for i, num in enumerate(nums):
            if target - num in nums_map and i != nums_map[target - num]:
                # nums_map[target - num] = index
                # n in dict = key ì¡°íšŒ   ,  ìžê¸° ìžì‹ ì„ ê³„ì‚°í•˜ëŠ” ê²½ìš° ì œê±°
                return [nums.index(num), nums_map[target - num]]
        

input_data = [[[2,7,11,15], 9] , [[3,2,4], 6] , [[3,4,3] , 6]]
sol = Solution()
for data in input_data:
    print(sol.func(data))
```

# inì„ ì´ìš©í•œ íƒìƒ‰
Time complexity of in : O(n)  
Time complexity : O(n^2)

```python
from typing import *
class Solution():
    def func(self, data:List[int]) -> List[int]:
        nums, target = data[0], data[1]
        for i, n in enumerate(nums):
            complement = target - n
            
            if complement in nums[i+1:]:
                return [nums.index(n), nums[i+1:].index(complement) + (i+1)]

input_data = [[[2,7,11,15], 9] , [[3,2,4], 6] , [[3,3] , 6]]
sol = Solution()
for data in input_data:
    print(sol.func(data))
```

# Brute-Force Way

```python
from typing import *
class Solution():
    def func(self, data:List[int]) -> List[int]:
        nums, target = data[0], data[1]
        for i in range(len(nums)):
            for j in range(i+1, len(nums)):
                if nums[i] + nums[j] == target:
                    return [i, j]

input_data = [[[2,7,11,15], 9] , [[3,2,4], 6] , [[3,3] , 6]]
sol = Solution()
for data in input_data:
    print(sol.func(data))
```
